```python
import pandas as pd
import numpy as np
import jieba.analyse
```

```python
stopWords = []
with open('stopWords.txt', 'r', encoding='UTF-8') as file:
    for data in file.readlines():
        data = data.strip()
        stopWords.append(data)
```

```python
stoplst = [' ', '\xa0']
for words in stoplst:
    stopWords.append(words)
```

```python
with open('userdict.txt', 'r', encoding='UTF-8') as file:
    for data in file.readlines():
        data = data.strip()
        jieba.add_word(data)
```

```python
df_terms = pd.read_csv('customer.csv')
```

```python
df_terms.head()
```

```python
error_lst = []
terms=[]
for i in range(len(df_terms['評論'])):
    try:
        for j in list(jieba.cut(df_terms['評論'][i], cut_all = False)):
            if j not in stopWords:
                terms.append(j)
    except:
        error_lst.append([i, df_terms['評論'][i]])
```

```python
names = {}          
relationships = {}  
lineNames = []  
for i in range(len(df_terms['評論'])):
    try:
        poss = jieba.cut(df_terms['評論'][i], cut_all = False)
        lineNames.append([])
        for w in poss:
            if w not in stopWords:
                lineNames[-1].append(w)        
            if names.get(w) is None and w not in stopWords:    
                relationships[w] = {}            
    except:
        pass
```

```python
term_dic = dict()
for sentence in lineNames:
    for term in sentence:
        if term not in term_dic:
            term_dic[term] = {}
```

```python
for i in range(len(lineNames)):
    for term1 in term_dic:
        num = 0
        for term2 in lineNames[i]:
            if term1 == term2:
                num += 1
        term_dic[term1][df_terms['評論'][i]] = num
```

```python
TDM = pd.DataFrame.from_dict(term_dic)
TDM.h
```

```python
#算出關係矩陣
X = TDM.to_numpy()
out = X.transpose().dot(X)
out
```
